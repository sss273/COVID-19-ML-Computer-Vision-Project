```python
import pandas as pd
import os
import shutil
```

```python
FILE_PATH = '/Users/sarahsheikh/Desktop/covid-chestxray-dataset-master/metadata.csv'
IMAGES_PATH = '/Users/sarahsheikh/Desktop/covid-chestxray-dataset-master/images'
```

```python
df = pd.read_csv(FILE_PATH)
print(df.shape)
```

```python
df.head
```

```python
TARGET_DIR = '/Users/sarahsheikh/Covid19/CovidDataset/Covid'

if not os.path.exists(TARGET_DIR):
    os.mkdir(TARGET_DIR)
    print('Covid folder created')
```

```python
cnt = 0

for (i,row) in df.iterrows():
    if row["finding"]=="Pneumonia/Viral/COVID-19" and row['view']=='PA':
        filename = row["filename"]
        image_path = os.path.join(IMAGES_PATH, filename)
        image_copy_path = os.path.join(TARGET_DIR, filename)
        shutil.copy2(image_path, image_copy_path)
        #print("Moving image", cnt)
        cnt +=1

#print(cnt)
```
